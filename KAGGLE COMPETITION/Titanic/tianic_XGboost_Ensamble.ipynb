```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('train (1).csv')
from sklearn.model_selection import train_test_split
data.columns
```

```python
Y=data['Survived']
X=data.drop(['Survived','PassengerId','Ticket','Name','Cabin'],axis=1)
X.head()
```

```python
X=pd.get_dummies(X,drop_first=True)
X.head()
```

```python
# for i in range(len(X)):
#     if pd.isna(X['Age'][i]):
#         X['Age'][i]=X['Age'].mean()
# #nan=np.nan        
# print(X['Age'][5])

X_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.2,random_state=666)
```

```python
#fist let;s make a model using Xgboost, then build an enesamble model 
from sklearn.model_selection import GridSearchCV
import xgboost as xgb 
para={'objective':'reg:squarederror'}
parameter_grid={'n_estimators' :np.arange(20,101),'max_depth':range(2,11),'colsample_bytree': [0.3, 0.7],    }
xgb_model=xgb.XGBRFClassifier(n_estimator=10)
model=GridSearchCV(param_grid=parameter_grid,estimator=xgb_model,cv=4,verbose=3,scoring='accuracy')
train_X_matrix=xgb.DMatrix(X)
model.fit(X_train,y_train)


```

```python
model.best_score_
```

```python
model.score(x_test,y_test)

```
